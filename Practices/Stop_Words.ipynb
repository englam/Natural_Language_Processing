### stop word是nltk內建的函數，裡面包含一些比較對於內容有關的參數，例如 a , this , the ....

```python
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
len(stopwords.words('english'))
```

```python
def content_fraction(text):
    stop_words = stopwords.words('english')
    content = [w for w in text if w.lower() not in stop_words]
    print (content)
    return len(content) / len(text)
```

```python
content_fraction(nltk.corpus.reuters.words())
```

```python
test = ['This','is','Englam','englam','ee','hhhhaaa']
```

```python
content_fraction(test)
```

```python
stop_words = stopwords.words('english')
content = [w for w in test if w.lower() not in stop_words]
```

```python
content
```

### 把this, a, the...這些都刪掉，留下剩下的單字

```python
def content_fraction2(text):
    stop_words = stopwords.words('english')
    content = [w for w in text if w.lower() not in stop_words]
    return content
```

```python
content_fraction2(test)
```

```python
test1 = ['be','englam']
```

```python
content_fraction2(test1)
```
