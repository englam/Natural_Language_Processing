### generate_model , 排列出最常出現的字元

cfdist[word].max() 

這個用法是找出，特定字元，之後會出現最多頻率的字
ex


指定living : cfd['living'].max()
最常出現 : 'creature'

指定capture : cfd['creature'].max()
最常出現 : 'that'


```python
def generate_model(cfdist, word, num=15):
    for i in range (num):
        print ('i:',i)
        print ('word:',word)
        word = cfdist[word].max()
```
