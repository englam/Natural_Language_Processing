### udhr是世界人權宣言的Library

```python
from nltk.corpus import udhr
import nltk
```

```python
languages = ['Chickasaw','English','German_Deutsch']
cfd = nltk.ConditionalFreqDist((lang, len(word)) for lang in languages for word in udhr.words(lang + '-Latin1'))
```

```python
cfd.plot()
```
