### 用比較的方式，將不屬於內建的英文單字找出來， 例如找出錯字

```python
#w.lower 將所有的字變成小寫
#w.isalpha 是否全是字母， 如果有數字或空白其他的就判為False，全是字母的話則為True

def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    print (text_vocab)
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

### 停用詞，通常是沒用的字，例如 "a" "the"等比較不具意義的字

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
#濾掉不屬於停字詞的用法,算出有多少字還留下的比例
def content_fraction_percentage(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
```

```python
content_fraction_percentage(nltk.corpus.reuters.words())
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return content
```

```python
puzzle_letters = nltk.FreqDist('egivrvonl')
obligatory = 'r'
```

```python
wordlist = nltk.corpus.words.words()
```

```python
puzzle_letters
```

```python
#找出字的長度大於六，並包含obligatory裡面的字，再用nltk.FreqDist(w)把用做頻率分佈，這個頻率分佈的字必需要小於或等於puzzle_letters裡面的文字及數量
[w for w in wordlist if len(w) >=6 and obligatory in w and nltk.FreqDist(w) <=puzzle_letters]
```

### 找出男生與女生都有的名字

```python
names = nltk.corpus.names
```

```python
names.fileids()
```

```python
male_names = names.words('male.txt')
female_names = names.words('female.txt')
```

```python
[w for w in male_names if w in female_names]
```

```python
# name[-1] 找出每個名字的最後一個字母
# 用機率條件顯示男生與女生結尾的字母出現的次數
cfd_name = nltk.ConditionalFreqDist((fileid,name[-1])
                                   for fileid in names.fileids()
                                   for name in names.words(fileid))
```

```python
cfd_name.plot()
```
